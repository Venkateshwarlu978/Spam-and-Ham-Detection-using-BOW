Spam and ham using BOW


```python
import pandas as pd
message=pd.read_csv("spam[1].csv",encoding="latin1")
```

```python
message
```

```python
message=message.drop(["Unnamed: 2",	"Unnamed: 3",	"Unnamed: 4"],axis=1)
```

```python
message
```

```python
##Data Cleaning and Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(message)):
  review=re.sub('[^a-zA-z]',' ',message['v2'][i])
  review=review.lower()
  review=review.split()
  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

Create Bag of words

```python
X = message['v2']   # text messages
y = pd.get_dummies(message['v1'], drop_first=True).astype(int)
```

```python
# X and y
corpus = message['v2']
y = pd.get_dummies(message['v1'], drop_first=True).astype(int)

# Train-test split (correct)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    corpus, y, test_size=0.20, random_state=42
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
##for binary Bow enable binary=True
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
# 1. Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

X_train_cv = cv.fit_transform(X_train)
X_test_cv = cv.transform(X_test)

# 2. Model
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train_cv, y_train)

# 3. Prediction
y_pred = spam_detect_model.predict(X_test_cv)
```
